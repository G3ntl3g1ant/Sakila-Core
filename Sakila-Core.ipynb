```python
# Answer for question 2 
# Because the Comedy category in the data is labeled as '5', this is what 
# was used to find the comedy films
p = """SELECT *
FROM film
JOIN film_category
ON film.film_id = film_category.film_id 
WHERE film_category.category_id = '5';"""
pd.read_sql(p,engine)
```

```python
# Answer for question 3
# The actor's ID was labeled as '5'
t = """SELECT *
FROM film
JOIN film_actor
ON film.film_id = film_actor.film_id
WHERE film_actor.actor_id = '5';"""
pd.read_sql(t,engine)
```

```python
# Answer for question 4
actors = """SELECT first_name, last_name
FROM actor AS a
INNER JOIN
film_actor AS fa
ON a.actor_id = fa.actor_id
INNER JOIN film AS f
ON fa.film_id = f.film_id
WHERE f.title = 'Bingo Talented';"""
pd.read_sql(actors, engine)
```

```python
# Answer to question 5
h = """SELECT customer_id, amount
FROM payment
WHERE amount >
	2 * (SELECT AVG(amount)
	FROM payment);"""
pd.read_sql(h, engine)
```

```python
# Answer to number 6
num_payments = """SELECT customer.first_name, customer.last_name,
 (SELECT COUNT(payment_id)
 FROM payment
 WHERE customer.customer_id = payment.customer_id) AS num_payments
FROM customer
ORDER BY num_payments DESC
LIMIT 5;"""
pd.read_sql(num_payments, engine)
```
